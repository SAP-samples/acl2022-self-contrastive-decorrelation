```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import string
from matplotlib.cm import coolwarm
from matplotlib import cm
```

```python
names = ['Sup. SimCSE$^+$ \n(81.6)','Unsup. SimCSE\n(74.5)', 'Avg. BERT (56.7)', 'SBERT$^+$ (74.9)', 'SBERT-flow$^+$ (76.6)', 'SBERT-whitening$^+$ (77.0)','BERT-flow (66.6)', 'BERT-whitening (66.3)', '$\mathbf{SCD}$ (74.2)']
x = [-3.65, -2.6, -1.4, -3.0,-3.6,-3.8, -3.7, -3.8, -3.833]
y = [0.24, 0.24, 0.19, 0.19, 0.44, 0.5, 0.55, 0.64, 0.84]
c=  [81.6, 74.5, 56.7, 74.9, 76.6, 77.0, 66.6,66.3, 74.2]
s = [175]*9
df = pd.DataFrame({'x':np.asarray(x), 'y':np.asarray(y)}, 
                  index=names)
```

```python

fig, ax = plt.subplots()
fig.set_figheight(6)
fig.set_figwidth(10)
df.plot('x', 'y', kind='scatter', ax=ax)
cmap = cm.get_cmap('coolwarm')
s = ax.scatter(df['x'], df['y'], c=c, cmap=cmap,s=s)

plt.xlim([-4, 0])
plt.grid()
ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')
cb = plt.colorbar(s)
for k, v in df.iterrows():
    ax.annotate(k, v,
                xytext=(10,-5), textcoords='offset points',
                family='sans-serif', fontsize=14, color='darkslategrey')
plt.ylabel(r'alignment', fontsize=18)
plt.xlabel(r'uniformity', fontsize=18)
plt.savefig('alignment_vs_uniformity.pdf', dpi=300, bbox_inches='tight')
```
